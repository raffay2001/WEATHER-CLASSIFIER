```python
import numpy as np
import matplotlib.pyplot as plt

# Performance metrics for each algorithm
algorithms = [
    "Logistic Regression",
    "Gaussian Naive Bayes",
    "Stochastic Gradient Descent",
]
metrics = ["Accuracy", "Precision", "Recall", "F1 Score"]
values_lr = [0.804, 0.802, 0.804, 0.803]
values_gnb = [0.763, 0.765, 0.763, 0.765]
values_sgd = [0.796, 0.798, 0.796, 0.797]

# Set the width of each bar
bar_width = 0.2

# Create the bar positions
bar_positions_lr = np.arange(len(metrics))
bar_positions_gnb = bar_positions_lr + bar_width
bar_positions_sgd = bar_positions_gnb + bar_width

# Plotting the bars for each algorithm
plt.bar(bar_positions_lr, values_lr, width=bar_width, label="Logistic Regression")
plt.bar(bar_positions_gnb, values_gnb, width=bar_width, label="Gaussian Naive Bayes")
plt.bar(
    bar_positions_sgd, values_sgd, width=bar_width, label="Stochastic Gradient Descent"
)

# Set the x-axis labels
plt.xticks(bar_positions_gnb, metrics)
plt.ylim(0, 1)  # Set the y-axis range from 0 to 1

# Set the title and labels
plt.title("Comparison of Performance Metrics")
plt.xlabel("Metrics")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Performance metrics for each algorithm
algorithms = ['Decision Tree', 'K Nearest Neighbours', 'Random Forest']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
values_dt = [0.875, 0.877, 0.875, 0.876]
values_knn = [0.906, 0.906, 0.906, 0.906]
values_rf = [0.930, 0.930, 0.930, 0.930]

# Set the width of each bar
bar_width = 0.2

# Create the bar positions
bar_positions_dt = np.arange(len(metrics))
bar_positions_knn = bar_positions_dt + bar_width
bar_positions_rf = bar_positions_knn + bar_width

# Plotting the bars for each algorithm
plt.bar(bar_positions_dt, values_dt, width=bar_width, label='Decision Tree')
plt.bar(bar_positions_knn, values_knn, width=bar_width, label='K Nearest Neighbours')
plt.bar(bar_positions_rf, values_rf, width=bar_width, label='Random Forest')

# Set the x-axis labels
plt.xticks(bar_positions_knn, metrics)
plt.ylim(0, 1)  # Set the y-axis range from 0 to 1

# Set the title and labels
plt.title('Comparison of Performance Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.legend()

plt.show()
```
